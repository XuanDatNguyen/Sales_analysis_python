### Import lib & load data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path = "C:/Users/Administrator/Desktop/Learn_Data/Portfolio Project/Sale Analysis/"
data_frame = pd.read_csv(path + 'sales2019_3.csv')
data_frame.head()
```

### Task 2 Clean and  Preprocess data

#### Task 2.1 Merge 12 month data

```python
files = os.listdir(path)
data_frames = []
all_length = []

for file in files:
    if file.endswith('.csv'):
        file_path = path + file
        data_frame = pd.read_csv(file_path)
        data_frames.append(data_frame)
        result = pd.concat(data_frames)
        length = len(data_frame)
        all_length.append(length)
df = result
np.sum(all_length) 

df.to_csv('Sales_2019.csv', index=False)

```

```python
df.head()
```

### Task 2.2 Add 'Month' column

```python
 #df['Order Date'].str.slice(0, 2, 1)
df['Month'] = df['Order Date'].str[0:2]
df.head(10)
```

```python
print(set(df['Month']))
```

### Task 2.3 Remove 'NaN' and 'Or' value

```python
df = df.dropna(how='all')
print(set(df['Month']))
```

```python
df = df[df['Month'] != 'Or']
df.head()
```

### Task3 Reporting

### Task 3.1: Tháng nào có doanh số cao nhất và bằng bao nhiêu

```python
# tạo cột total_money, group by theo cột Month
```

```python
#conver dtype
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], downcast='integer')
df['Price Each'] = pd.to_numeric(df['Price Each'], downcast='float')


df['Total Money'] = df['Quantity Ordered'] * df['Price Each']
df.head()
```

```python
# chuyển col total money sang cạnh order date
moving_col = df.pop('Total Money')
df.insert(4, 'Total Money', moving_col)
df.head()
```

```python
#tháng có doanh thu cao nhất
max_revenue = df.groupby('Month').sum()['Total Money'].max()

max_revenue
```

```python
revenue_by_month = df.groupby('Month').sum()['Total Money']
months = np.linspace(1,12,12, dtype="int")
months
```

```python
plt.bar(months, revenue_by_month);
plt.title("Doanh thu theo tháng")
plt.xticks(months)
plt.xlabel("Tháng")
plt.ylabel("Doanh thu (Triệu $)");
```

### Task 3.2: Thành phố nào có doanh thu cao nhất và bằng bao nhiêu?

### Tách cột City

```python

#address = '944 Walnut St, Boston, MA 02215'
#address.split(',')[1]

#def getCityName(address):
#   return address.split(',')[1]

address_to_city = lambda address:address.split(',')[1]

df['City'] = df['Purchase Address'].apply(address_to_city)
df.head()
```

```python
#groupby theo city
revenue_by_city = df.groupby('City').sum()['Total Money']
#Doanh thu cao nhất theo thành phố
revenue_by_city.max()
```

```python
#cities = []
#for city, revenue in revenue_by_city.items():
#    cities.append(city)
    
#Comprehension
#cities = [city for city, revenue in revenue_by_city.items()]
    
cities = df['City'].unique()
```

```python
plt.bar(cities, revenue_by_city)
plt.title('Doanh thu theo thành phố')
plt.ylabel('Doanh thu (Triệu $)')
plt.xticks(rotation=45)
plt.xlabel('Thành phố')
```

### Doanh nghiệp cần chiếu quảng cáo vào khung giờ nào để tăng khả năng mua hàng của khách hàng.

```python
df.head()
```

```python
#convert order date sang dtype datetime
df["Order Date"] = pd.to_datetime(df["Order Date"])
df['Hours'] = df['Order Date'].dt.hour
df.head()
```

```python
#khung giờ nào bán được nhiều hàng nhất
revenue_by_hours = df.groupby('Hours').count()['Order ID']
revenue_by_hours.max()
```

```python
#Comprehension
hours = [hour for hour, revenue in revenue_by_hours.items()]
    
print(hours)
```

```python
plt.plot(hours, revenue_by_hours)
plt.grid()
plt.title("Số lượng đơn hàng theo khung giờ")
plt.ylabel("Số đơn hàng")
plt.xlabel("khung giờ")
plt.xticks(hours, size=8);
```

### Task 3.5: sản phẩm nào được bán nhiều nhất

```python
df.head()
```

```python
product_qty = df.groupby('Product').count()['Quantity Ordered']
product_avg_price = df.groupby('Product').mean()['Price Each']
products = []

for product, qty in product_qty.items():
    products.append(product)
    
product_qty.max()
```

```python
plt.bar(products, product_qty)
plt.plot(products, product_avg_price)
plt.title('Số lượng bán ra theo sản phẩm')
plt.xlabel('Sản phẩm')
plt.ylabel('Số lượng sản phẩm')
plt.xticks(products, rotation=90, size=8);
```

```python
x = products
y1 = product_qty
y2 = product_avg_price
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(x, y1, color='g')
ax2.plot(x, y2, 'b-')

ax1.set_xlabel('Sản phẩm')
ax1.set_ylabel('Số lượng', color='g')
ax1.set_xticklabels(products, rotation=90, size=8)
ax2.set_ylabel('Giá bán', color='b');

```
